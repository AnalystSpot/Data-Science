```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn import svm
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
from sklearn.metrics import confusion_matrix
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import KFold
k_fold = KFold(n_splits=10, shuffle=True, random_state=0)

path = '/content/drive/MyDrive/My_Projects/finalyYearProj/inal/covid.csv'
df = pd.read_csv(path)
display(df)

df.replace(('Yes', 'No'), (1, 0), inplace=True)
display(df)

### Split data to train & test

x=df.drop(['Id','COVID-19'],axis=1)
y=df['COVID-19']
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0)

# Model Building

#random forest
rfc = RandomForestClassifier().fit(x_train, y_train)
acc=rfc.score(x_test, y_test)
rfc1=round(acc,6)
print('Algorithm : RandomForestClassifier , Accuracy',rfc1)

#logistic regression
lr = LogisticRegression().fit(x_train, y_train)
acc_lr=lr.score(x_test, y_test)
lr1=round(acc_lr,6)
print('Algorithm : LogisticRegression , Accuracy',lr1)

#Decision tree
DTC = DecisionTreeClassifier().fit(x_train, y_train)
acc_dtc=DTC.score(x_test, y_test)
dtc1=round(acc_dtc,6)
print('Algorithm : DecisionTreeClassifier , Accuracy',dtc1)

#support vaector machine
svc = svm.SVC().fit(x_train, y_train)
acc_svc=svc.score(x_test,y_test)
svm1=round(acc_svc,6)
print('Algorithm : svm , Accuracy',svm1)

#naive bayes
gnb=GaussianNB().fit(x_train,y_train)
acc_gnb=gnb.score(x_test,y_test)
gnb1=round(acc_gnb,6)
print('Algorithm : GaussianNB , Accuracy',gnb1)

y_pred=rfc.predict(x_test)
cm=confusion_matrix(y_test, y_pred)
sns.heatmap(cm,fmt='d',annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')

# Predict for new values

display(x_test)

display(x_test.iloc[11].values)

display(y_pred[11])

display(rfc.predict([[1, 1, 0, 1, 0, 0, 1, 0, 1, 1, 1, 0, 0, 1, 0, 0, 0, 0]]))

# Model tuning function

# best model tuning - gridsearchcv \ Lasso , DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import ShuffleSplit

def find_best_model_using_grid_search_cv(x,y):
    algos={
        'Random Forest Classifier':{
            'model':RandomForestClassifier(),
            'params':{
                "max_depth":[15,20,25], 
                "n_estimators":[27,30,33],
                "criterion":["gini", "entropy"]
            }
        },
        'Logistic Regression':{
            'model':LogisticRegression(),
            'params':{
                "penalty":['l1', 'l2', 'elasticnet', 'none'],
                "solver":['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
            }
        },
        'Decision Tree Calssifier':{
            'model':DecisionTreeClassifier(),
            'params':{
                "criterion":["gini", "entropy"],
                "max_depth":[15,20,25]
            }
        },
        'SVM':{
            'model':svm.SVC(),
            'params':{
                'C': [0.1, 1.0, 10, 100],
                'gamma': [0.1, 1.0, 10, 100]
            }
        },
        'Naive Bayes':{
            'model':GaussianNB(),
            'params':{
                'var_smoothing': np.logspace(0,-9, num=100)
            }
        }
    }
    
    scores=[]
    cv=ShuffleSplit(n_splits=5,test_size=0.2,random_state=0)
    for algo_name,config in algos.items():
        gs=GridSearchCV(config['model'],config['params'],
                       cv=cv,
                        return_train_score=False
                       )
        gs.fit(x,y)
        scores.append({
            'model':algo_name,
            'best_score':gs.best_score_,
            'best_params':gs.best_params_
        })
    return pd.DataFrame(scores,columns=['model','best_score','best_params'])

display(find_best_model_using_grid_search_cv(x_train,y_train))

import pickle
pickle.dump(rfc,open('covid_prediction.pickle','wb'))
```
